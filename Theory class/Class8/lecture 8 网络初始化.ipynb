### 6个全连接层，使用高斯随机数进行网络初始化，tanh为激活函数，查看每一层的输出分布

```python
import numpy as np
import matplotlib.pyplot as plt

dims=[4096]*7
hs= []
x=np.random.randn(16,dims[0])
for Din, Dout in zip(dims[:-1],dims[1:]):
    print(Din,Dout)
    W= 0.01 * np.random.randn(Din, Dout)
    x= np.tanh(x.dot(W))
    hs.append(x)
    

num_plots = len(hs)
# Create a figure with subplots
fig, axs = plt.subplots(nrows=1,ncols=num_plots, figsize=(num_plots*4,4))

# Iterate over each column and plot the histogram
for i in range(num_plots):
    column_data = np.array(hs)[i].flatten()
    ax = axs[i] if num_plots > 1 else axs
    ax.hist(column_data, bins='auto')
    ax.set_xlim([-1, 1])  # Set the x-axis limit
    

    ax.set_xlabel('Value')
    ax.set_ylabel('Frequency')
    ax.set_title('Histogram of Layer {}'.format(i+1))

# Adjust the layout and display the plot
plt.tight_layout()
plt.show()
```

### 6个全连接层，使用高斯随机初始化，Relu为激活函数，查看每一层的输出分布

```python
import numpy as np
import matplotlib.pyplot as plt

dims=[4096]*7
hs= []
x=np.random.randn(16,dims[0])
for Din, Dout in zip(dims[:-1],dims[1:]):
    print(Din,Dout)
    W= 0.01 * np.random.randn(Din, Dout)
    x= np.maximum(0,x.dot(W))
    hs.append(x)
    

num_plots = len(hs)
# Create a figure with subplots
fig, axs = plt.subplots(nrows=1,ncols=num_plots, figsize=(num_plots*4,4))

# Iterate over each column and plot the histogram
for i in range(num_plots):
    column_data = np.array(hs)[i].flatten()
    ax = axs[i] if num_plots > 1 else axs
    ax.hist(column_data, bins='auto')
    ax.set_xlim([-.1, .1])  # Set the x-axis limit
    

    ax.set_xlabel('Value')
    ax.set_ylabel('Frequency')
    ax.set_title('Histogram of Layer {}'.format(i+1))

# Adjust the layout and display the plot
plt.tight_layout()
plt.show()
```
